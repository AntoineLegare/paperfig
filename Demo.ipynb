# Loading / generating synthetic data

```python
x = np.linspace(0, 10 * 2 * np.pi, 10000)
y = np.sin(x)
```

# Rendering figure

```python
fig = PaperFigure(figsize=(7, 3), dpi=300)

# Global parameters --------------------------------------------------------------------------------

fig.set_tick_length(2)
fig.set_font_size(6)
fig.add_background()

# Setting up axes ----------------------------------------------------------------------------------

# Four small axes
w = 1.25
pad = (7 - 4 * w) / 3
for i in range(4):
    fig.add_axes(f'scatter{i}', (i * (w + pad), 0), w, w)

# Large rectangle
fig.add_axes('curve', (0, 1.75), 7, 1.5)

fig.set_line_thickness(0.5) # Must be after axes have been made

# Filling axes -------------------------------------------------------------------------------------

ax = fig.axes['scatter0']
ax.scatter(np.random.normal(0, 1, 100), np.random.normal(0, 1, 100), s=5, color='black')
ax.spines[['top', 'right']].set_visible(False)

ax = fig.axes['scatter1']
ax.scatter(np.random.normal(0, 1, 100), np.random.normal(0, 1, 100), s=5, color='red')
ax.spines[['top', 'right']].set_visible(False)

ax = fig.axes['scatter2']
ax.scatter(np.random.normal(0, 1, 100), np.random.normal(0, 1, 100), s=5, color='blue')
ax.spines[['top', 'right']].set_visible(False)

ax = fig.axes['scatter3']
ax.scatter(np.random.normal(0, 1, 100), np.random.normal(0, 1, 100), s=5, color='green')
ax.spines[['top', 'right']].set_visible(False)

ax = fig.axes['curve']
ax.plot(x, y, color='black')
ax.set_xlim([0, x[-1]])
ax.spines[['top', 'right']].set_visible(False)

# Displaying ---------------------------------------------------------------------------------------

fig.show()
```
